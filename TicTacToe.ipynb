```python
from IPython.display import clear_output

def display_board(board):
    
    clear_output()
    
    print (board[1],'|',board[2],'|',board[3])
    print('----------')
    print (board[4],'|',board[5],'|',board[6])
    
    game_on = False

while True:
    # Initialize the board
    board = ['#', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']

    # Display the board
    display_board(board)

    print('Welcome to Tic Tac Toe game coded by Tevfik')

    # Get player markers (e.g., X or O)
    (player1_marker, player2_marker) = player_input()

    # Determine the order of players
    (player_first, player_second) = choose_first(player1_marker, player2_marker)

    game_on = True

    while game_on:
        # Player 1's turn
        position = player_choice(board)
        place_marker(board, player_first, position)

        if win_check(board, player_first):
            print('Congrats! Player 1 is the winner')
            break
        else:
            pass

        if full_board_check(board):
            print('The game is a tie!')
            break
        else:
            pass

        # Player 2's turn
        position = player_choice(board)
        place_marker(board, player_second, position)

        if win_check(board, player_second):
            print('Congrats! Player 2 is the winner')
            break
        else:
            pass

        if full_board_check(board):
            print('The game is a tie!')
            break
        else:
            pass

    game_on = False

    if not replay():
        break

    print('----------')
    print (board[7],'|',board[8],'|',board[9])
    
```

```python
def win_check(board, mark):
    # Check for winning combinations horizontally, vertically, and diagonally
    if board[1] == board[2] == board[3] == mark:
        return True
    elif board[4] == board[5] == board[6] == mark:
        return True
    elif board[7] == board[8] == board[9] == mark:
        return True
    elif board[1] == board[4] == board[7] == mark:
        return True
    elif board[2] == board[5] == board[8] == mark:
        return True
    elif board[3] == board[6] == board[9] == mark:
        return True
    elif board[3] == board[5] == board[7] == mark:
        return True
    elif board[1] == board[5] == board[9] == mark:
        return True
    else:
        return False
```

```python
import random

def choose_first(player1, player2):
    # Generate a random number (either 1 or 2) to determine the starting player
    a = random.randint(1, 2)
    
    # Check the randomly generated number and determine the starting player accordingly
    if a == 1:
        print(f'The game is starting with {player1}')
        return (player1, player2)
    else:
        print(f'The game is starting with {player2}')
        return (player2, player1)

    
    
```

```python
def space_check(board, position):
    # Check if the specified position on the board is empty
    if board[position] == ' ':
        return True
    else:
        return False
```

```python
def full_board_check(board):
    # Check if the board is full by iterating through positions 1 to 9
    for i in range(1, 10):
        # If there is an empty space on the board, return False
        if board[i] == ' ':
            return False
            break
        else:
            i += 1
    
    # If all positions are filled, return True
    else:
        return True
```

```python
def player_choice(board):
    # Prompt the player to pick a position on the board
    position = int(input('Please pick a number (1-9): '))
    
    # Validate the chosen position
    while space_check(board, position) != True:
        print('Please pick another number that has not been chosen!')
        position = int(input('Please pick a number (1-9): '))
    
    # Return the validated position chosen by the player
    return position
```

```python
def replay():
    # Ask the user if they want to replay the game
    answer = input('Do you want to replay? (Y or N): ')
    
    # Validate the user's input
    while answer != 'Y' and answer != 'N':
        print('Please type Y or N!')
        answer = input('Do you want to replay? (Y or N): ')
        
    # Return True if the user wants to replay, False otherwise
    if answer == 'Y':
        return True
    else:
        return False
```
